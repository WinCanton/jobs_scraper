```python
soup = BeautifulSoup(r.content, "lxml")
results = soup.findAll('div', {"class":"col-xs-12 job-results clearfix"})
results
```

```python
results[0].find_all('div', {"class":"page-title"})
```

```python
def get_job_title(parsed_text):
    jobs = parsed_text[0].find_all('div', {"class": "job-title"})
    job_title = []
    for job in jobs:
        job_title.append(job.contents[1].contents[1].contents[0])
    return job_title
```

```python
get_job_title(results)
```

```python
len(get_job_title(results))
```

```python
def get_job_type(parsed_text):
    jobs = parsed_text[0].find_all('li', {"class":"job-type"})
    job_types = []
    for job in jobs:
        job_types.append(job.contents[1].contents[0])
    return job_types
```

```python
get_job_type(results)
```

```python
len(get_job_type(results))
```

```python
def get_job_salary(parsed_text):
    jobs = parsed_text[0].find_all('li', {"class":"salary"})
    job_salary = []
    for job in jobs:
        job_salary.append(job.contents[0])
    return job_salary
```

```python
get_job_salary(results)
```

```python
len(get_job_salary(results))
```

```python
def extract_text(input_text):
    if input_text:
        return input_text
    else:
        return 'None'
```

```python
extract_text('Martin')
```

```python
extract_text('')
```

```python
def get_job_company(parsed_text):
    jobs = parsed_text[0].find_all('li', {"class":"company"})
    job_company = []
    for job in jobs:
        try:
            job_company.append(extract_text(job.a['title']))
        except:
            job_company.append('None')
    return job_company
```

```python
get_job_company(results)
```

```python
len(get_job_company(results))
```

```python
def get_job_recruiter(parsed_text):
    jobs = parsed_text[0].find_all('li', {"class":"company"})
    job_recruiter = []
    for job in jobs:
        try:
            job_recruiter.append(extract_text(job.a['title']))
        except:
            job_recruiter.append('None')
    return job_recruiter
```

```python
def get_job_post_date(parsed_text):
    jobs = parsed_text[0].find_all('li', {"class":"date-posted"})
    job_post_date = []
    for job in jobs:
        job_post_date.append(job.span.text.strip())
    return job_post_date
```

```python
get_job_post_date(results)
```

```python
def get_job_location(parsed_text):
    jobs = parsed_text[0].find_all('li', {"class":"location"})
    job_location = []
    for job in jobs:
        job_location.append(job.span.text.strip())
    return job_location
```

```python
get_job_location(results)
```
